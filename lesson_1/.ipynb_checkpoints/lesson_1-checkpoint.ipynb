````python
df2['Date Posted'] = pd.to_datetime(df2['Date Posted'])
df2['Date Collected'] = pd.to_datetime(df2['Date Collected'])
df2.dtypes
````

````python
for col in columns:
    print(col)
    print(df2[col].unique())
````

````python
df2.loc[:, columns] = df2.loc[:, columns].applymap(lambda x: np.nan if x == 'Non-detect' else x)
df2.loc[:, columns] = df2.loc[:, columns].applymap(lambda x: np.nan if x == 'ND' else x)
````

````python
for col in columns:
    df2[col] = pd.to_numeric(df2[col])
````

## Export data in different format

````python
url = 'https://raw.githubusercontent.com/htetaunglynn94/big_data_analysis_python/refs/heads/main/lesson_1/dataset/RadNet_Laboratory_Analysis.csv'
df = pd.read_csv(url)
df.head()
````

````python
columns = df.columns
id_cols = ['State', 'Location', "Date Posted", 'Date Collected', 'Sample Type', 'Unit']
rmd_cols = list(set(columns) - set(id_cols))
rmd_cols
````

````python
df['Cs-134'] = df['Cs-134'].apply(lambda x: np.nan if x == "Non-detect" else x)
df.loc[:, columns] = df.loc[:, columns].applymap(lambda x: np.nan if x == 'Non-detect' else x)
df.loc[:, columns] = df.loc[:, columns].applymap(lambda x: np.nan if x == 'ND' else x)
````

````python
df.head()
````

````python
target_col = ['State', 'Location', 'Sample Type', 'Unit']
````

````python
df[target_col] = df[target_col].applymap(lambda x: str(x).strip())
````

````python
df.head()
````

````python
df['Date Posted'] = pd.to_datetime(df['Date Posted'])
df['Date Collected'] = pd.to_datetime(df['Date Collected'])
````

````python
for col in rmd_cols:
    df[col] = pd.to_numeric(df[col])
````

````python
df['State'] = df['State'].astype('category')
df['Location'] = df['Location'].astype('category')
df['Unit'] = df['Unit'].astype('category')
df['Sample Type'] = df['Sample Type'].astype('category')
````

````python
df.to_csv('radiation_clean.csv', index=False, sep=';', encoding='utf-8')
````

````python
df.to_parquet('radiation_clean.prq', index=False)
````
