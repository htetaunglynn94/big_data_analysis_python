# Exercise 2: Missing Value Preprocessing with High Reproducibility

````python
df.head()
````

````python
import collections, random
import numpy as np
````

````python
replaced = collections.defaultdict(set)
replaced
````

<details><summary>Example of `collections.defaultdict()` usage</summary>  


**With integer**   

```python
d = collections.defaultdict(int)  # Default value is 0 for non-existent keys
print(d['a'])                     # Outputs: 0 (instead of KeyError)
d['b'] += 1
print(d['b'])                     # Outputs: 1
```   


**With list**  

```python
d = collections.defaultdict(list)
d['a'].append(1)
print(d['a'])                     # Outputs: [1]
```  

    
**With set**  
    
```python
d = collections.defaultdict(set)  # Default value is an empty set
d['a'].add(1)
print(d['a'])                     # Outputs: {1}
```  
</details>

````python
# 
ix = [(row, col) for row in range(df.shape[0]) for col in range(df.shape[1])]
ix[0:5]
````

````python
# check for number of rows and columns mapping dataframe dimension
len(set([r for r,c in ix])), len(set([c for r,c in ix])), df.shape, len(df)
````

**Note**: `df.shape[0]` = `len(df)`

````python
len(ix), len(ix) * .1, round(len(ix) * .1)
````

````python
# 10% of the total cells will be replaced
to_replace = int(round(.1 * len(ix)))
to_replace
````

````python
for row, col in ix:                         # iterate for all row and columns
    if len(replaced[row]) < df.shape[1]-1:  # if number of replaced row is less than total rows
        df.iloc[row, col] = np.nan          # fill NaN value to the random location of the dataframe
        to_replace -= 1                     # count down to_replace to know how many items leave
        replaced[row].add(col)              # add value to a dict key (e,g, {1095: {0}})
        if to_replace == 0:                 # if to_replace value is equal to zero
            break                           # break the operation
````

#### Test the code

````python
# Create a sample DataFrame
d = pd.DataFrame({
                    'Col 0': [1, 6, 11], 'Col 1': [2, 7, 12],
                    'Col 2': [3, 8, 13], 'Col 3': [4, 9, 14], 
                    'Col 4': [5, 10, 15]})

display(d)
print("Total cells:", d.size)

# Initialize the replaced dictionary as a defaultdict of sets
r_placed = collections.defaultdict(set)
print(r_placed)

# Create a list of all (row, column) indices
x = [(row, col) for row in range(d.shape[0]) for col in range(d.shape[1])]
print("x values:", x)

# Shuffle the indices randomly
random.shuffle(x)
print("random no:", x)

# Calculate how many values to replace (10% of total cells)
t_replace = int(round(0.1 * len(x)))

# Randomly replace values in the DataFrame with NaN
for row, col in x:
    # Print statements for debugging
    print(f"Processing row: {row}, col: {col}")
    print(f"Replaced columns in row {row}: {r_placed[row]}")
    
    # Ensure we do not replace all columns in a row
    if len(r_placed[row]) < d.shape[1] - 1:
        d.iloc[row, col] = np.nan  # Replace the value with NaN
        t_replace -= 1  # Decrease the number of replacements left
        r_placed[row].add(col)  # Mark this column as replaced
        
        # Break the loop if we have reached the desired number of replacements
        if t_replace == 0:
            break

# Print the final DataFrame and the replaced dictionary
print("\nFinal DataFrame:")
print(d)
print("\nReplaced dictionary:")
print(r_placed)
````
